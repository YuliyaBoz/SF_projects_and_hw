```python
#игроки
players = {'x':'Крестики', 'o':'Нолики'}

#размерность матрицы NxN
N = 3
cells = N**2

#создаем пустое игровое поле
field = [['-' for i in range(N)] for j in range(N)] 


#функция вывода поля:
def print_field(matrix):
    print('-------------')
    print(' ', *range(N))
    for i in range(N):
        print(i, end=' ')
        for j in range(N):
            print(matrix[i][j], end=' ')
        print()
    print('-------------')
    print()

    
#Функция ввода ячейки координат клетки игроком и проверок на корректный ввод
def input_move(tac):
    in_move = input(f'{players[tac]}! Введите номер строки и номер столбца ячейки, куда вы хотите поставить {tac}, через пробел: \n')
    
    if not in_move.replace(' ', '').isdigit():
        print('Не все значения числовые')
        return input_move(tac)
    
    moves = list(map(int, in_move.split()))
    
    if len(moves) != 2:
        print('Вы ввели неверное количество чисел или забыли пробел.')
        return input_move(tac)
    
    if not all(move in range(N) for move in moves):
        print('Вы ввели значение за пределами поля')
        return input_move(tac)
        
    return moves
    

# Проверка, свободна ли клетка для хода:
def true_input(matrix, i, j):
    return (matrix[i][j] == '-')

        
#Перебираем все вариации выигрышных сочетаний:
def win_combination(matrix, tac):
    tic_or_tac = [tac for i in range(N)]

    for i in range(N):
        row = matrix[i]
        if tic_or_tac == row:
            return [(i,j) for j in range(N)]  
   
    for j in range(N):
        column = []
        for i in range(N):
            column.append(matrix[i][j])
        if tic_or_tac == column:
            return [(i,j) for i in range(N)]
    
    diagonal_1 = []
    for i in range(N):
        diagonal_1.append(matrix[i][i])
    if tic_or_tac == diagonal_1:
        return [(i,i) for i in range(N)]
    
    diagonal_2 = []
    for i in range(N):
        diagonal_2.append(matrix[i][N-1-i])
    if tic_or_tac == diagonal_2:
        return [(i,N-1-i) for i in range(N)]


#Игра
def game_tic_or_tac():
    print_field(field)
    
    cells = N**2
    tic = 'x'
    while cells:
        move = input_move(tic)
        
        if true_input(field, move[0], move[1]):
            field[move[0]][move[1]] = tic 
        else:
            print('Ячейка занята. Выберите другую')
            continue
            
        win_comb = win_combination(field, tic)
        if win_comb is not None:
            print(f'{players[tic]}! Поздравляю, вы победили!')
            for (i, j) in win_comb:
                field[i][j] = field[i][j].upper()
            print_field(field)
            break
            
        print_field(field)
        tic = 'o' if tic == 'x' else 'x'
        cells -= 1
        
    else:
        print('Ничья! Сыграйте еще разок')

        
game_tic_or_tac()   
```
